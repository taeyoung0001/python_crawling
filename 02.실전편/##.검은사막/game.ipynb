```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []
for i in range(1, 10):
    # URL 정의 (루프 내에서 매번 업데이트)
    URL = f'https://m.inven.co.kr/board/black/3584?p={i}'
    
    response = requests.get(URL)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    items = soup.select('.li-wrap')

    for item in items:
        # 링크, 제목, 카테고리 추출
        link = item.select_one(".li-wrap > a").attrs['href']
        title = item.select_one(".subject").text.strip()
        category = item.select_one(".in-cate").text.strip()

        # 추천 수 처리
        reco_element = item.select_one(".reco")
        
        if reco_element and reco_element.text.strip():
            # "추천" 문자열을 기준으로 추천 수 추출
            recommand_text = reco_element.text.strip()
            if "추천" in recommand_text:
                recommand = recommand_text.split("추천")[1].strip()
            else:
                recommand = recommand_text.strip()  # 추천이라는 텍스트 없이 그냥 숫자만 있을 경우
        else:
            # 추천 수가 없으면 빈 문자열 처리
            recommand = ""  

        print(title, link, category, recommand)
        # 데이터 리스트에 추가
        data.append([title, link, category, recommand])

# DataFrame 생성
df = pd.DataFrame(data, columns = ['제목', '링크', '카테고리', '추천 수'])

# Excel로 저장
df.to_excel("game.xlsx", index=False)
```
